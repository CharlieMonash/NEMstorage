```python
import requests

# URL of the CSV file
#url = 'https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_202301_NSW1.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
# Send a GET request to the URL
states = ['NSW1', 'VIC1', 'QLD1', 'SA1', 'TAS1']
#states = ['NSW1']
for state in states:
    for year in range(1998, 2024):
        for month in range(1, 13):
        # Construct the URL for the current year and month
            year = str(year)
            if month< 10:
                month = str(month)
                month = '0'+month
            else:
                month = str(month)
            url = f'https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_'+year+month+'_'+state+'.csv'
        


            response = requests.get(url, headers=headers)

# Check if the request was successful
            if response.status_code == 200:
    # Save the contents of the response to a file
                with open('PRICE_AND_DEMAND_'+year+month+'_'+state+'.csv', 'wb') as f:
                    f.write(response.content)
                print('https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_'+year+month+'_'+state+'.csv File downloaded successfully!')
            else:
                print(f'Error downloading file https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_'+year+month+'_'+state+f'.csv. Status code: {response.status_code}')
```
